```python
# Scatter plot 

    --> A scatter plot is a diagram where each value in the data set is representd by a dot...
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter([1,2,3,5],[6,7,8,9])
plt.show()
```

```python
plt.scatter([1,2,3,5],[6,7,8,9],color='red')
plt.show()
```

```python
# Area plot
--> Filling the area between lines

--> legend()

    * legend creates a legend with descriptive lables for each plotted data series
        
                            (or)
        
    * legend is used to specify the location of the legend...
```

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]
plt.stackplot(days,working,sleeping,colors=['r','g'],labels=['work','rest'])
plt.legend()
plt.show()
```

```python
# Pie chart
   
    * A pie() function that generates a pie diagram represents data in an array..
    * Pie chart are best to use when you are trying to compare parts of a whole..
```

```python
marks = [23,56,78,11,56]
marks1 = [34,89,88,45]
names = ['c','java','c++','python','oracle']
plt.pie(marks,labels = names)
plt.legend()
```

```python
# plotting two or more graphs on same plot
```

```python
# scatter plot model
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]
plt.scatter(x1,y1,label='line1') # plotting the line 1 points

# line 2 points
x2 = [6,7,8]
y2 = [8,3,9]

plt.scatter(x2,y2,label='line2') # plotting the line 2 points

plt.xlabel('x-axis') # naming the x- axis
plt.ylabel('y-axis') # naming the y-axis
plt.title("Two lines  in same graph") # giving a title to graph

plt.legend() # show a legend on the plot
plt.show() # function to show the plot
```

```python
# line plot model
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]
plt.plot(x1,y1,label='line1') # plotting the line 1 points

# line 2 points
x2 = [6,7,8]
y2 = [8,3,9]

plt.plot(x2,y2,label='line2') # plotting the line 2 points

plt.xlabel('x-axis') # naming the x- axis
plt.ylabel('y-axis') # naming the y-axis
plt.title("Two lines  in same graph") # giving a title to graph

plt.legend() # show a legend on the plot
plt.show() # function to show the plot
```

```python
# scatter plot
x = [5,6,7,2,3,4,8,9,1] # x-axis values
y = [4,3,7,8,2,1,5,6,9] # y-axis values

plt.scatter(x,y,label="stars",color="green",marker='*')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("MY SCATTER PLOT")
plt.legend()

plt.show()
```

```python
# customization of plots

x = [1,2,3,4,5,6]
y = [2,4,1,5,2,6]
plt.plot(x,y,color='red',linestyle='dashed',linewidth=3,marker='o',markerfacecolor='green',markersize=12)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("some cool customization")
plt.show()
```

```python
# dashdot style
x = [1,2,3,4,5,6]
y = [2,4,1,5,2,6]
plt.plot(x,y,color='red',linestyle='dashdot',linewidth=3,marker='o',markerfacecolor='green',markersize=12)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("some cool customization")
plt.show()
```

```python
# solid style
x = [1,2,3,4,5,6]
y = [2,4,1,5,2,6]
plt.plot(x,y,color='red',linestyle='solid',linewidth=3,marker='o',markerfacecolor='green',markersize=12)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("some cool customization")
plt.show()
```
